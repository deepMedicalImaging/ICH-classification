```r
ICH <- read.csv("/content/Age_and_sex - Hoja 1.csv")
```

```r
summary(ICH)
```

```r
ICH$SEX <- factor (ICH$SEX)
ICH$HIC <- factor (ICH$HIC)
ICH$AGE <- as.numeric (ICH$AGE)
```

```r
summary(ICH)
```

```r
# SEX

tabla_1 <- table(ICH$HIC, ICH$SEX)
tabla_1

prop.table(tabla_1, margin = 1)

barplot(tabla_1, legend = TRUE)
```

```r
#FISHER'S EXACT TEST (we cannot assume normality to perform a chi square test):

fisher.test(ICH$HIC, ICH$SEX)
```

```r
# AGE

hist(ICH$AGE, main = "Age distribution", xlab = "Age (years)")
```

```r
hist(log(ICH$AGE))
```

```r
plot(ICH$HIC, ICH$AGE, xlab = "Diagnose (0-normal, 1-ICH)", ylab = "Age (years)", main = "Age distribution")
```

```r
# Mann-Whitney U test (because we cannot assume normality to use a T test):

 wilcox.test(AGE~HIC,data=ICH)
```
