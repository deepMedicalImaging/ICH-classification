```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import keras
from math import sqrt
import sklearn.metrics
import scipy
```

```python
array_of_train_images= np.load('path to the x_train_images_array.npy')
labels_train_array= np.load('path to the y_train_labels_array.npy')
array_of_test_images= np.load('path to the x_test_images_array.npy')
labels_test_array= np.load('path to the y_test_labels_array.npy')


print(array_of_train_images.shape)
print(labels_train_array.shape)
print(array_of_test_images.shape)
print(labels_test_array.shape)
```

```python
predictions = np.load('path to the predictions of the model.npy')
predictions.shape
```

```python
def confusion_matrix (preds, y_test, threshold):

    tps=[]
    fps=[]
    fns=[]
    tns=[]

    for i in range (preds.shape[0]):
        pred= 0 if preds[i][0] < threshold else 1
        if pred == 1 and y_test[i] == 1:
          tps.append(i)
        if pred == 1 and y_test[i] == 0:
          fps.append(i)
        if pred == 0 and y_test[i] == 1:
          fns.append(i)
        if pred == 0 and y_test[i] == 0:
          tns.append(i)
    
    return tps, fps, fns, tns
```

```python
tps, fps, fns, tns= confusion_matrix (predictions, labels_test_array, 0.9)
```

```python
#VISUALIZING WRONG FLAGGED IMAGES
def panel (images, group, columns, title):
  
  nrows= len(group)//columns
  if len(group)%columns > 0:
    nrows= nrows+1
  fig,axs = plt.subplots(nrows,columns, figsize=(15,15))
  fig.suptitle(title, fontsize=25, y=0.92)
  counter=-1
  for i in range (columns):
    for j in range(nrows):
      counter=counter+1
      if counter >= len(group):
        continue
      else:
        index= group[counter]
        axs[i,j].imshow(images[index])
      axs[i,j].set_xticks([])
      axs[i,j].set_yticks([])
  plt.subplots_adjust(wspace=0, hspace=0.05)

  fig.savefig(title + '.jpg', dpi=500, bbox_inches = 'tight')
```

```python
panel (array_of_test_images, fns, 3, 'Images of False Negatives (all)')
```

```python
panel (array_of_test_images, fps, 4, 'Images of False Positives (all)')
```
